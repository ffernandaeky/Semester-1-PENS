```python
%matplotlib inline
plt.xlabel('co(karbon)')
plt.ylabel('no2(nitrogen)')
plt.scatter(df.co,df.no2, color='red')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['co']],df.co)
```

```python
reg.predict([[20]])
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
%matplotlib inline
plt.xlabel('co(karbon)')
plt.ylabel('no2(nitrogen)')
plt.scatter(df.co,df.no2, color='red')
plt.plot(df.co,reg.predict(df[['co']]), color='blue')
```
